```python
# Setting up engine for voice assistant
engine = pyttsx3.init('sapi5')
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[1].id)

# send_message function is called whenever user types a prompt and hence, it returns a response
def send_message(event):
  # Get the user's input
  prompt = input_field.get()

  # Clear the input field
  input_field.delete(0, tk.END)

  # Generate a response from the chatbot
  response = get_response(prompt)

  # Display the response in the chatbot's text area
  text_area.insert(tk.END, f"User: {prompt}\n")
  text_area.insert(tk.END, f"Chatbot: {response}\n")

# speak function speaks the response whenever user ask question using microphone
def speak(audio):
    engine.say(audio)
    engine.runAndWait()

# take_command function listens the prompt spoken by the user
def take_command():
     
    recognizer = sr.Recognizer()
    
    with sr.Microphone() as source:
        audio = recognizer.listen(source)
    
    try:   
        prompt = recognizer.recognize_google(audio, language ='en-in')
  
    except Exception as e:
        text_area.insert(tk.END, f"Chatbot: Unable to Recognize your voice.\n") 
        return None 
     
    response = get_response(prompt)
    
    text_area.insert(tk.END, f"User: {prompt}\n")
    text_area.insert(tk.END, f"Chatbot: {response}\n")
    
    speak(response)

# Creating window for interface
window = tk.Tk()
window.title("Amena")
window.config(bg='lavender')
text_area = tk.Text(window, bg="lavender")
text_area.grid(row=1)

# Create the user's input field
input_field = tk.Entry(window, width=50, font=('Arial', 16))
input_field.grid(row=2, column=0)

# sending prompt
window.bind('<Return>', send_message)

# Microphone button
btn = tk.Button(text='Mic', height = 2, width = 2, command=take_command)
btn.grid(row=2, column=1)

window.mainloop()
```
